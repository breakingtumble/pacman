```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3]
y = [0.7, 0.4, 0.2]

plt.plot(x, y) 
  
# naming the x axis 
plt.xlabel('Number of ghosts') 
# naming the y axis 
plt.ylabel('Win rate') 
  
# giving a title to my graph 
plt.title('Win rate - Number of ghosts relationship') 
  
# function to show the plot 
plt.show() 
```

```python
# Значення Win Rate для глибини відповідно на карті minimaxClassic
x1 = [1, 2, 3, 4, 5]
y1 = [0.7, 0.7, 0.75, 0.85, 0.90]

plt.plot(x1, y1, label = "Win Rate on minimaxClassic") 

# Значення Win Rate для глибини відповідно на карті smallClassic

x2 = [1, 2, 3, 4, 5]
y2 = [0.55, 0.7, 0.75, 0.8, 0.7]

plt.plot(x2, y2, label = "Win Rate on smallClassic") 

plt.title('Win rate - Depth relationship') 
  
plt.legend() 
   
plt.show() 
```

```python
# Значення Average score для глибини відповідно на карті minimaxClassic
x1 = [1, 2, 3, 4, 5]
y1 = [210.85, 209.9, 258.15, 360.7, 412.4]

plt.plot(x1, y1, label = "Win Rate on minimaxClassic") 

# Значення Average score для глибини відповідно на карті smallClassic

x2 = [1, 2, 3, 4, 5]
y2 = [476.65, 713.45, 862.05, 862.95, 880.75]

plt.plot(x2, y2, label = "Average Score on smallClassic") 

plt.title('Average score - Depth relationship') 
  
plt.legend() 
   
plt.show()
```

```python
# Значення Win Rate для глибини відповідно на карті smallClassic з функцією betterEval
x1 = [1, 2, 3, 4, 5]
y1 = [0.55, 0.7, 0.75, 0.8, 0.7]

plt.plot(x1, y1, label = "Win Rate on smallClassic without betterEval") 

# Значення Win Rate для глибини відповідно на карті smallClassic з функцією betterEval

x2 = [1, 2, 3, 4, 5]
y2 = [0.3, 0.6, 0.6, 0.6, 0.75]

plt.plot(x2, y2, label = "Win Rate on smallClassic with betterEval") 

plt.title('Win rate - Depth relationship') 
  
plt.legend() 
   
plt.show() 
```
